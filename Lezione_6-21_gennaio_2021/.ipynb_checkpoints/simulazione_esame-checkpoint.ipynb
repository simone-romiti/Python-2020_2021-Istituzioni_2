```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 20, 10000)
r = t

x = r*np.cos(t)
y = r*np.sin(t)
```

```python
fig = plt.figure()

ax = fig.gca()

ax.plot(x,y)

plt.show()
```

```python
eps = 0.1

xm, ym = [], []

n = len(x)

for i in range(0, n):
    if y[i]<eps and y[i]>-eps:
        xm.append(x[i])
        ym.append(y[i])
        
print(xm, ym)
```

```python
fig = plt.figure()

ax = fig.gca()

ax.plot(xm,ym, marker="*", color="blue", linestyle="None")
ax.plot(x,y, color="red")

plt.show()
```

```python
# z = x^2 + 3*y^2

x_val = np.linspace(-10, 10, 100)

%matplotlib notebook
xs, ys =  np.meshgrid(x_val, x_val)
zs = xs**2 + 3*ys**2

fig = plt.figure()

ax = fig.gca(projection="3d")

ax.set_ylim(-10,10)
ax.set_ylim(-10,10)
ax.set_zlim(0, 30)

ax.plot_surface(xs,ys,zs, color="red")

plt.show()

```
