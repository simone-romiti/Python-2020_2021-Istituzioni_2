```python
# importo le librerie che servono
import sympy

# definisco i vettori
v1 = sympy.Matrix([1,2,3])
v2 = sympy.Matrix([0,-3,9])

# faccio il prodotto scalare
s = float(v1.dot(v2))
print("Prodotto scalare tra", v1, "e", v2, " = ", s)

# calcolo l'angolo compreso tra i due
```

```python
import numpy as np

def to_vec(v_s):
    v = []
    for vi in v_s:
        v.append(float(vi))
    return v

def norm(v_s):
    v = to_vec(v_s)
    return np.sqrt( v[0]**2 + v[1]**2 + v[2]**2)

  
n1, n2 = norm(v1), norm(v2)
cs = s/(n1*n2)  # cos(theta)
theta = np.arccos(cs)
print("the angle between the two vectors is", theta, "radians")
```

```python
a = -1
b = 2
c = -4
d = 8
```

```python
# %matplotlib notebook

ti = np.linspace(-5,5, 100)

xl = a*ti
yl = b*ti
zl = c*ti - d/c

t,s = np.meshgrid(ti, ti)

# ax + by + cz + d = 0
xp = t
yp = s
zp = -(a*xp + b*yp + d)/c
```

```python
fig = plt.figure()
ax = fig.gca(projection="3d")

ax.set_xlim(-20,20)
ax.set_ylim(-20,20)
ax.set_zlim(-20,20)

ax.plot(xl,yl,zl)
ax.plot_surface(xp,yp,zp)

plt.show()
```

```python
print("Intersection")

x,y,z = sympy.symbols("x,y,z")
```

```python
# find the intersection between the plane and the line perpendicular and passing through the origin

el1 = y - b*x/a
el2 = z - c*x/a
ep = a*x + b*y + c*z + d

display(el1)
display(el2)
display(ep)
```

```python
s1 = sympy.solve(ep, z)
display(s1[0])

s2 = el2.subs(z, s1[0])
display(s2)

s3 = sympy.solve(s2, y)
display(s3[0])

x_its = float(sympy.solve(s3[0], x)[0])
y_its = float(sympy.solve(s2, y)[0].subs(x, x_its))
z_its = float(s1[0].subs(x, x_its).subs(y, y_its))

print(x_its, y_its, z_its)

```

```python
r_sph = np.sqrt(x_its**2 + y_its**2 + z_its**2)
r_sph2 = np.fabs(d)/np.sqrt(a**2 + b**2 + c**2)

print(r_sph, r_sph2)

theta_vals = np.linspace(0, np.pi)
phi_vals = np.linspace(0, 2*np.pi)

theta, phi = np.meshgrid(theta_vals, phi_vals)

x_sph = r_sph*np.sin(theta)*np.cos(phi)
y_sph = r_sph*np.sin(theta)*np.sin(phi)
z_sph = r_sph*np.cos(theta)

ti = np.linspace(-5,5, 100)

xl = a*ti
yl = b*ti
zl = c*ti - d/c

t,s = np.meshgrid(ti, ti)

# ax + by + cz = d
xp = t
yp = s
zp = -(1/c)*(a*xp + b*yp + d)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 5))
ax = fig.gca(projection="3d")

r_max = 5
ax.set_xlim(-1.3*r_max, 1.3*r_max)
ax.set_ylim(-1.3*r_max, 1.3*r_max)
ax.set_zlim(-r_max, r_max)


ax.plot_surface(xp, yp, zp, color="red", alpha=0.3)
ax.plot_surface(x_sph, y_sph, z_sph, alpha=0.3)
ax.plot(xl, yl, zl)
ax.plot(x_its, y_its, z_its, marker="*")


plt.show()
```
