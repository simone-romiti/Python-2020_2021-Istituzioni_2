```python
def plot3d_quadric(eq, x, a, b, y, c, d, z):
    sol = solve(eq, z)
    for s in sol:
        display(s)
    plot3d(sol[0], sol[1], (x,a,b), (y,c,d))
```

Step per disegnare l'iperboloide come superficie rigata

* funzione per n punti su circ.
* funzione per retta tra due punti
* ciclo for su tutti i punti della circonferenza

```python
import numpy as np
def circ_points(n, phi, h):
    pts = []
    zc = h
    for i in range(0,n):
        theta = phi + (2*np.pi/n)*i
        xc = cos(theta)
        yc = sin(theta)
        pts.append([xc, yc, zc])
    return pts

print("check")
print(circ_points(10, 0.0, 3))
```

```python
def get_plot_line(A0, B0, I=[-5, 5]):
    A = np.array(A0)
    B = np.array(B0)
    v = A-B
    t = symbols("t")
    rt = A + v*t
    return plot3d_parametric_line(rt[0], rt[1], rt[2], (t,I[0], I[1]), show=False)

```

```python
print("test")
p1 = get_plot_line([0.0, 0.0, 0.0],[1.0, 3.0, -4.0], [-10, 10])
p1.show()
```

```python
P_hyp = sympy.plotting.plot(show=False)
```

```python
N = 40
phi = 0.6
h = 1
c1 = circ_points(N, 0.0, 0.0)
c2 = circ_points(N, phi, h)

for i in range(0, N):
    P_hyp.extend(get_plot_line(c1[i], c2[i]))

P_hyp.show()
```
