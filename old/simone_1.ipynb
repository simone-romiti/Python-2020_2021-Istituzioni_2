## sympy library

Now we explore the functionalities of sympy

```python
from sympy import * # we import the entire sympy
from sympy.plotting import * 
```

```python
x, y, z = symbols('x y z') # defining symbols
```

### solution of algebraic equations

```python
print("We can solve algebraic equations")
S1 = solve(x**2-4, x)
print(S1)
for s in S1:
  display(Eq(x,s))
```

```python
eqn = Eq(x**2-y+1, 0)
print("We want to solve the following equation:")
display(eqn)
print("The solutions are:")
S2 = solve(eqn,x)
for s in S2:
  display(Eq(x,s))
```

### Plotting

Sympy can be used for plotting

```python
print("Paraboloids:")
z0 = 10
PRB = z0 + sqrt(x**2+y**2)
z1 = PRB
z2 = -PRB
display(Eq(z, z1),Eq(z, z2))

L = 100
plot3d(z1,z2, (x,-L,L),(y,-L,L))
```

```python
def plot3d_quadric(eq,x,a,b,y,c,d,z):
    sol = solve(eq,z) # solving for 'z'
    print("Plotting the following curves:")
    for s in sol:
      display(Eq(z,s))
    plot3d(sol[0],sol[1],(x,a,b),(y,c,d))
```

```python
plot3d_quadric(x**2+y**2+z**2-16,x=x,a=-5,b=5,y=y,c=-5,d=5,z=z)
```

```python
Q1 = z**2-x**2-y**2
display(Eq(Q1, 0))
plot3d_quadric(z**2-x**2-y**2,x,-5,5,y,-5,5,z)
```

```python
Q2 = x**2+y**2-z**2-1
display(Eq(Q2, 0))
print("Hyperboloid from the 2 z(x,y) curves:")
plot3d_quadric(Q2,x,-5,5,y,-5,5,z)
```

#### Parametric plots

We now consider the hyperboloid as a [ruled surface](https://en.wikipedia.org/wiki/Ruled_surface).
The equation is:

$$ \frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1$$

the parametrized version is:

$$ x(t,s) = a \sqrt{1 + t^2} \, \cos{s}$$
$$ y(t,s) = b \sqrt{1 + t^2} \, \sin{s}$$
$$ z(t,s) = c t$$


for simplicity we use $a=b=c=1$.


```python
t,s = symbols('t, s')
a=b=c=1
from numpy import pi as pi
n=20
print("That's your hyperbolid!")
plot3d_parametric_surface(a*sqrt(1+t**2)*cos(s),b*sqrt(1+t**2)*sin(s), c*t, (t, -5, 5), (s,0,2*pi))
```

```python
plot3d_parametric_line(a*t,b*t, c*t, (t, -5, 5))
```
