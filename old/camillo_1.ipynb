```python
def pari_o_dispari(x):
    if type(x)==int and x>=0 and x%2==0:
        print(x," è pari")
    elif type(x)==int and x>=0 and x%2!=0:  # elif significa else if, cioè altrimenti se
        print(x," è dispari")
    else:
        print(x," non è un numero naturale")
```

```python
pari_o_dispari(5)
```

```python
x, y, z = symbols('x y z')
```

```python
solve(x**2-y+1,x)
```

```python
plot3d(sqrt(x**2+y**2),-sqrt(x**2+y**2), (x,-5,5),(y,-5,5))
```

```python
def plot3d_quadric(eq,x,a,b,y,c,d,z):
    sol=solve(eq,z)
    return plot3d(sol[0],sol[1],(x,a,b),(y,c,d))
```

```python
plot3d_quadric(x**2+y**2+z**2-16,x,-5,5,y,-5,5,z)
```

```python
plot3d_quadric(z**2-x**2-y**2,x,-5,5,y,-5,5,z)
```

```python
# %matplotlib notebook
plot3d_quadric(x**2+y**2-z**2-1,x,-5,5,y,-5,5,z)
```
