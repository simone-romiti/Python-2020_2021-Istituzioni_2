```python
# CALCOLO DIFFERENZIALE E INTEGRALE
```

```python
x, y, z, t = symbols('x y z t')  # con questo camando, x,y e z non sono più variabili qualsiasi ma delle variabili "simboliche"
```

```python
sin(x)**2+cos(x)**2
```

```python
simplify(sin(x)**2+cos(x)**2)
```

```python
solve(x**2-x, x) # equazioni algebriche
```

```python
diff(sin(x) + x**2, x) # derivata
```

```python
integrate(cos(x),x)  # integrale indefinito
```

```python
integrate(cos(x), (x,0,pi/2))  # integrale definito
```

```python
# definisco una funzione
```

```python
def fun(x):
    return 1 + sin(x - pi/2)
```

```python
plot(fun(x), (x,-4,4))
```

```python
#%matplotlib notebook
plot3d(y*sin(x), (x,-5,5),(y,-5,5))
```

```python
#%matplotlib notebook
plot3d(sqrt(x**2+y**2),-sqrt(x**2+y**2), (x,-5,5),(y,-5,5))
```

```python
plot_implicit(Eq(x**2+y**2,16),(x,-5,5),(y,-5,5))
```

```python
plot_implicit(Eq(x**2+y**2,16),(x,-5,5),(y,-5,5),aspect_ratio=(1,1))
```

```python
plot_parametric(cos(t),sin(t),(t,0,2*pi),aspect_ratio=(1,1))
```
