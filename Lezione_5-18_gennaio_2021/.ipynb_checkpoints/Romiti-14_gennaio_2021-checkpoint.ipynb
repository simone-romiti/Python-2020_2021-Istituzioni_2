```python
import numpy as np
import sympy
from sympy import *
from sympy.plotting import *
x, y, z = symbols("x, y, z")
```

```python
display(sin(x)*(x**3 + x - 5)/(x**2 + 4))

def f(x):
    return sin(x)*(x**3 + x - 5)/(x**2 + 4)
```

```python
a, b = -4,4
p0 = plot(f(x), (x, a, b), line_color="blue",show=False)
p1 = plot(diff(f(x), x), (x,a,b), line_color="red", show=False)
p2 = plot( diff(  diff(f(x), x),  x  )  , (x,a,b), line_color="magenta", show=False)

p0.extend(p1)
p0.extend(p2)

p0.show()
```

```python
theta = symbols("\\theta")
display(theta)
```

```python
rt_sym = 1 - cos(theta)
xt_sym = rt_sym*cos(theta)
yt_sym = rt_sym*sin(theta)

display(xt_sym)
display(yt_sym)
```

```python
import matplotlib.pyplot as plt

theta_vec = np.linspace(0, 2*np.pi, 100) # 100 punti tra 0 e 2*pi
print(theta_vec)
```

```python
x_vec = (1-np.cos(theta_vec))*np.cos(theta_vec)
y_vec = (1-np.cos(theta_vec))*np.sin(theta_vec)

plt.plot(x_vec, y_vec)
plt.show()
```

```python
plt.axes().set_aspect("equal") # stessa scala per x e y
plt.plot(x_vec, y_vec) # plot dell'intera cardioide

n = 8
for i in range(0, n): # plot di alcuni punti individualmente
    theta_i = i*(2*np.pi/n)
    # prima rimpiazzo theta con theta_i e poi trasformo il tutto in numero con la virgola
    xi = xt_sym.subs(theta, theta_i).evalf() 
    yi = yt_sym.subs(theta, theta_i).evalf()
    plt.plot(xi, yi, marker="^", color="red")

plt.show()
```

```python
def get_segment(A, B, n=100): # punti sul segmento AB in 2D
    xi = np.linspace(A[0], B[0])
    yi = np.linspace(A[1], B[1])
    return [xi, yi]
```
