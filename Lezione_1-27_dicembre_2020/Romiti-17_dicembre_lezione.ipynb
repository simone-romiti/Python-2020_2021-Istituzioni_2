```python
from sympy import * # we import the entire sympy
from sympy.plotting import * 
```

```python
x, y, z = symbols("x, y, z") # definisco x,y,z
display(x)
display(y)
display(z)
```

```python
e1 = pow(x,2)*y**5 - x*y +1
display(e1)
```

```python
ls = solve(e1, x)
display(ls[0])
display(ls[1])
```

```python
a, b, c = symbols("a, b, c")
p2 = a*x**2 + b*x + c
L2 = solve(p2, x)
display(L2[0])
display(L2[1])
```

```python
f1 = exp(x)*sin(x)/log(x)
display(f1)
der1 = diff(f1, x)
display(der1)
```

```python
f2 = exp(x)*sin(x)
Inte1 = Integral(f2, x)
display(Inte1)
display(Inte1.doit())
```

```python
v = Matrix([x, y, z])
w = Matrix([3, 1, 2])

display(v+w)
display(v[1,0])
```

```python
a = Matrix([1,0,0])
b = Matrix([0,1,1])
display(a.dot(b)) # dot product (prodotto scalare)
```

```python
a = Matrix([1,0,0])
b = Matrix([0,1,1])
display(a.cross(b)) # cross product (prodotto vettoriale)
```

```python
a = Matrix([1,0,0])
b = Matrix([0,1,1])
c = Matrix([3,4,-3])

z = b.cross(c)
display(z)

display(a.dot(z))

# display(a.dot(b.cross(c)))
```
